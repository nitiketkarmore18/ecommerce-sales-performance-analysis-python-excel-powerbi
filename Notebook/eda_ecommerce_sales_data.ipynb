```
import pandas as pd

df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

```
df[['Sales','Profit','Quantity','Discount']].describe()
```

```
print("Total Sales:", df['Sales'].sum())
print("Total Profit:", df['Profit'].sum())
print("Profit Margin:", df['Profit'].sum() / df['Sales'].sum())
```

```
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.show()
```

```
df.groupby('Category')['Profit'].sum().plot(kind='bar')
plt.title("Profit by Category")
plt.show()
```

1. Category Performance Insight

🔹 Observation:

Sales and profit vary significantly across product categories.

🔹 Insight:

High sales in certain categories do not necessarily translate into high profit, indicating inefficiencies in pricing strategy and discount allocation.

🔹 Business Action:

Optimize pricing models and review discount strategies for low-profit categories that still generate high sales to improve overall profitability.

```
df.groupby('Sub-Category')['Profit'].sum().sort_values().plot(kind='barh')
plt.title("Profit by Sub-Category")
plt.show()
```

2. Sub-Category Performance Insight

🔹 Observation:

Certain sub-categories are generating low or negative profit despite contributing to overall sales.

🔹 Insight:

This indicates that some products are being sold at a loss or with very low margins, impacting overall business profitability.

🔹 Business Action:

Reassess pricing, reduce excessive discounting, and consider discontinuing or redesigning consistently loss-making sub-categories.

```
df.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales by Region")
plt.show()
```

4. Region Performance Insight

🔹 Observation:

Sales distribution is uneven across regions, with certain regions contributing significantly more revenue than others.

🔹 Insight:

A few regions are dominating revenue generation, indicating imbalance in market penetration and sales performance.

🔹 Business Action:

Strengthen marketing and distribution strategies in underperforming regions while maintaining strong performance in high-revenue regions.

```
df.groupby('Segment')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Segment")
plt.show()
```

5. Segment Performance Insight

🔹 Observation:

The Consumer segment contributes the highest share of total sales compared to Corporate and Home Office segments.

🔹 Insight:

The business is highly dependent on individual consumers, indicating a strong B2C-driven revenue model.

🔹 Business Action:

Continue focusing on the Consumer segment while exploring growth opportunities in Corporate and Home Office segments to diversify revenue streams.

```
df.groupby('Ship Mode')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Ship Mode")
plt.show()
```

6. Shipping Mode Insight

🔹 Observation:

Different shipping modes show variation in sales distribution, with Standard Class being the most commonly used.

🔹 Insight:

Customers prefer standard shipping due to cost efficiency, while premium shipping options are less frequently used.

🔹 Business Action:

Optimize logistics for Standard Class shipping and evaluate pricing strategy for faster shipping modes to increase adoption without hurting margins.

```
import seaborn as sns

sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```

3. Discount vs Profit Insight

🔹 Observation:

There is a noticeable variation in profit with increasing discount levels, and higher discounts often correspond to reduced or negative profit.

🔹 Insight:

Excessive discounting is negatively impacting profitability. While discounts may drive higher sales volume, they are significantly reducing overall margins.

🔹 Business Action:

Introduce a controlled discounting strategy with defined thresholds. Avoid high discounts on already low-margin products and analyze discount impact before applying promotional offers.
